```python
import pandas as pd
import numpy as np
df = pd.read_csv('./육회test.csv',encoding = 'utf-8-sig')
```

```python
print(df)
df.describe()
```

```python
df = df.rename(columns={'Unnamed: 0':'랭킹'})
df = df.drop(df[(df['스마트스토어'] == 0)].index)
df = df.dropna(subset=['리뷰'])
df = df.drop(df[(df['랭킹'] > 50) & (df['리뷰'] < 30)].index)
```
